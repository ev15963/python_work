```python
# 폴더 내 파일 하나씩 읽기 --> 합치기
delete_list = ['아시아주','미주','구주','대양주','아프리카주','교포소계','계','전년동기','성장률(%)','구성비(%)']

df = pd.DataFrame()

for file in os.listdir(folder):
    fpath = folder + '/' + file
    temp = pd.read_excel(fpath, header = 1)
    
    yearmonth = file.split('.')[0].split('_')[1] 
    temp['기준년월'] = yearmonth
    
    cond = [ ]
    for nation in temp['국적']:
        if nation in delete_list:
            cond.append(False)
        else:
            cond.append(True)
    temp = temp[ cond ]    

    
    df = df.append(temp, ignore_index = True)
```

```python
#nation=country_top[1]

for nation in country_top:

    #nation
    cond = df['국적'] == nation
    df_filter = df[ cond ]
    df_filter.head()


    # 예제 3-42 시계열 그래프 그리기 2(꾸미기)
    ## 그래프 크기 조절 
    plt.figure(figsize = (12, 4))

    ## df_filter 그래프 내용 설정 
    plt.plot(df_filter['기준년월'], df_filter['관광'])

    ## 그래프 타이틀, X축, Y축 이름 달기 
    plt.title('중국 국적의 관광객 추이'.format(nation))
    plt.xlabel('기준년월')
    plt.ylabel('관광객수')

    ## x 축 눈금 값 설정 
    plt.xticks(['201001', '201101', '201201', '201301', '201401', '201501', '201601', '201701', '201801', '201901'])

    ## 그래프 표현하기 
    plt.show()
```

```python
# 여러 국가 그래프 그리기

for nation in country_top:

    cond = df['국적'] == nation
    df_filter = df[ cond ]

    ## 그래프 크기 조절 
    plt.figure(figsize = (12, 4))

    ## 그래프 내용 설정 
    plt.plot(df_filter['기준년월'], df_filter['관광'])

    ## 그래프 타이틀, X축, Y축 이름 달기 
    plt.title('{} 국적의 관광객 추이'.format(nation))
    plt.xlabel('기준년월')
    plt.ylabel('관광객수')

    ## x 축 눈금 값 설정 
    plt.xticks(['201001', '201101', '201201', '201301', '201401', '201501', '201601', '201701', '201801', '201901'])

    ## 그래프 표현하기 
    plt.show()
```
