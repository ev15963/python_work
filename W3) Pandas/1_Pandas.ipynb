## 2.1.5. 데이터 통합하기① - 옆으로 통합(pd.merge)

```python
# 예제 2-21 데이터 확인하기 
sample_1
```

```python
# 예제 2-22 국적코드표 엑셀파일 불러오기 
code_master = pd.read_excel('./files/sample_codemaster.xlsx')
code_master
```

```python
pd.merge(sample_1, code_master,
        how = 'left',
        left_on =  '국적코드',
        right_on = '국적코드') # 'right', 'inner', 'outter'
```

```python
pd.merge(sample_1, code_master,
        how = 'inner',
        left_on =  '국적코드',
        right_on = '국적코드') # 'right', 'inner', 'outter'
```

```python
pd.merge(sample_1, code_master,
        how = 'right',
        left_on =  '국적코드',
        right_on = '국적코드') # 'right', 'inner', 'outter'
```

```python
pd.merge(sample_1, code_master,
        how = 'outer',
        left_on =  '국적코드',
        right_on = '국적코드') # 'right', 'inner', 'outter'
```

```python
# 예제 2-23 데이터 옆으로 통합하기(left 조건) 
sample_1_code = pd.merge(left=sample_1, 
                         right=code_master,
                         how='left',
                         left_on='국적코드',
                         right_on='국적코드')
sample_1_code
```

```python
sample_1_code = sample_1_code[ ['국적코드', '성별', '입국객수', '기준년월', '국적명']] #col 순서 바꿀때
```

```python
# 예제 2-24 데이터 옆으로 통합하기(inner 조건) 
sample_1_code_inner = pd.merge(left=sample_1, 
                               right=code_master,
                               how='inner',
                               left_on='국적코드',
                               right_on='국적코드')
sample_1_code_inner
```
