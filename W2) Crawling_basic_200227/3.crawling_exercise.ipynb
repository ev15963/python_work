## 1. 라이브러리 호출

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

## 2. 크롬드라이버 열고, html 정보 가져오기

```python
browser = webdriver.Chrome('./driver/chromedriver.exe')
```

```python
url = 'http://www.genie.co.kr/chart/top200'
browser.get(url)
```

```python
html=browser.page_source
html
```

```python
soup=BeautifulSoup(html, "html.parser")
soup
```

### 7.2. 반복문으로 Top50 의 각 노래 정보를 리스트로 만들기

```python
result = []

for element in lst:
    singer = element.select('a.artist.ellipsis')[0].text
    album=element.select('a.albumtitle.ellipsis')[0].text
    title=element.select('a.title.ellipsis')[0].text.replace('\n','').replace(' ', '')

    song_info = [title, singer, album]
    
    result.append(song_info)
```

```python
result
```

### 7.3. Top50 데이터 프레임으로 만들기 

```python
result_df=pd.DataFrame(result)
```

```python
result_df
```

```python
result_df.columns=['제목','가수','앨범명']
result_df
```
